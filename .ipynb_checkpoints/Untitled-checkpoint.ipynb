```python
from matplotlib import pyplot
from matplotlib.image import imread
 
folder= "train/"

for i in range(9):
    pyplot.subplot(330 +1 + i)
    filename= folder +"cat/"+ 'cat' + str(i) +".jpg"
    img= imread(filename)
    pyplot.imshow(img)
    
pyplot.show()
```
